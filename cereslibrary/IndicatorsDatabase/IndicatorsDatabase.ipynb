### Location retrieval module

```python
latitude = 40
longitude = -98
```

```python
input_point = np.array([float(latitude), float(longitude)]) 

#input_point = np.array([39.014908, -98.010465])

input_point_df = pd.DataFrame(
    {'Name': ['CAFO1'],
     'Latitude': [input_point[0]],
     'Longitude': [input_point[1]]})

#    poly  = geopandas.GeoDataFrame.from_file('cereslibrary/GIS/watershed/huc8sum.shp')
poly  = geopandas.GeoDataFrame.from_file('watershed/huc8sum.shp')
point = geopandas.GeoDataFrame(input_point_df, geometry=geopandas.points_from_xy(input_point_df.Longitude, input_point_df.Latitude))

poly.crs = {'init' :'epsg:4326'}
point.crs = {'init' :'epsg:4326'}

pointInPolys = sjoin(point, poly, how='left')

#HUC8ContPoint = float(pointInPolys['HUC_8'])
HUC8ContPoint = pointInPolys['HUC_8'].values[0]

HUC8ContPoint
```

```python
P_TES_indicators_df = pd.read_csv('P_TES_indicators_df.csv', converters={'HUC_8': lambda x: str(x)}).fillna(0)
SoilGeoChemical_indicators_df = pd.read_csv('SoilGeoChemical_indicators_df.csv', converters={'HUC_8': lambda x: str(x)}).fillna(0)
TSI_indicators_df = pd.read_csv('TSI_indicators_df.csv', converters={'HUC_8': lambda x: str(x)}).fillna(0)
```

```python
IndicatorsDatabase_i = pd.merge(P_TES_indicators_df, SoilGeoChemical_indicators_df, on="HUC_8", how='outer')
IndicatorsDatabase = pd.merge(IndicatorsDatabase_i, TSI_indicators_df, on="HUC_8", how='outer')
IndicatorsDatabase = IndicatorsDatabase.fillna('No Data')
IndicatorsDatabase
```

```python
# =============================================================================
# GIS data retrieval
# =============================================================================
GIS_data = IndicatorsDatabase.set_index('HUC_8').loc[HUC8ContPoint]
GIS_data
```
